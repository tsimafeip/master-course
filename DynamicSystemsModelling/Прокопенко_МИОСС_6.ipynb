### № 9.1.5

**Решение:**
Необходимо построить отображение из R в R с устойчивым 3-циклом, проходящим через x=0. По теореме Шарковского непрерывное отображение с 3-циклом влечет за собой хаос при изменении начальных параметров, а в условии не сказано, что отображение должно быть непрерывным.
Поэтому для простоты построим кусочно-гладкое отображение с устойчивым 3-циклом, проходящим через x=0<br> 

Например:
$$f(x) = \begin{cases} 
0, & \mbox{при } x\mbox{$\in(-\infty; -1]$} \\
1, & \mbox{при } x\mbox{$\in$(-1; 1)} \\
-2, & \mbox{при } x\mbox{$\in$[1; +$\infty$)}
\end{cases}
$$

Подтвердим ответ теоретически, численно и графически.<br>
1) Теоретически.<br>
По определению ясно, что f(x) - это 3-цикл, где $x^{*}_1=0, x^{*}_2=1, x^{*}_3=-2$, т.к. f(f(f($x^{*}_i$)))=$x^{*}_i$.<br>
Осталось доказать устойчивость. Найдем производную f(x).
Так как эта функция кусочно-гладкая, то на любом из трех интервалов существует производная.
f'($x^{*}_i$)=0 на всех интервалах, т.к. значения f(x) не зависят от x.

И abs(f'($x^{*}_1$)*f'($x^{*}_2$)*f'($x^{*}_3$))<1 => функция имеет устойчивый цикл.

2) Численно.
Выберем несколько различных начальных значений. Например, -3, 0, 2.<br>
a) x_1=f(-3)=0. 
f(0) = 1
f(1) = -2
f(-2) = 0.
<br>
б) x_1=f(0)=1.
f(1) = -2
f(-2) = 0
f(0) = 1.
<br>
в) x_1=f(2)=-2. 
f(-2) = 0
f(0) = 1
f(1) = -2.
